```python
import matplotlib.pyplot as plt

# Create the bar plot
ax = df['accommodates'].value_counts().plot(kind='bar')

# Add data labels on top of the bars
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))

# Show the plot
plt.show()
```
